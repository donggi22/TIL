````python
# 양의 정수 n의 팩토리얼 구하기

def factorial(n: int) -> int:
    """양의 정수 n의 팩토리얼값을 재귀적으로 구함"""
    if n > 1:
        return n * factorial(n - 1)
    else:
        return 1

if __name__ == '__main__':
    n = int(input('출력할 팩토리얼값을 입력하세요.: '))
    print(f'{n}의 팩토리얼은 {factorial(n)}입니다.')
````

````python
# 유클리드 호제법으로 최대 공약수 구하기

def gcd(x: int, y: int) -> int:
    if y == 0:
        return x
    else:
        return gcd(y, x % y)
    
if __name__ == '__main__':
    print('두 정숫값의 최대 공약수를 구합니다.')

x = int(input('첫번째 정숫값을 입력하세요.: '))
y = int(input('두번째 정숫값을 입력하세요.: '))

print(f'두 정숫값의 최대 공약수(gcd({x}, {y}))는 {gcd(x, y)}입니다.')
````

````python
# 피보나치 수열 구하기

def fibonacci(n: int) -> int:
    if n <= 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input('구하고 싶은 피보나치 수열의 항 번호를 입력하세요: '))

print(f'{n + 1}번째 피보나치 수열의 항은 {fibonacci(n)}입니다.')
````

````python
# 순수한 재귀 함수 구현하기

def recur(n: int) -> int:
    """순수한 재귀 함수 recur의 구현"""
    if n > 0:
        recur(n - 1)
        print(n)
        recur(n - 2)

x = int(input('정수값을 입력하세요.: '))

recur(x)
````

````python
# recur() 함수의 재귀 호출을 거꾸로 출력하기

def recur(n: int) -> int:
    """순수한 재귀 함수 recur의 구현(거꾸로 출력)"""
    if n > 0:
        recur(n - 2)
        print(n)
        recur(n - 1)

x = int(input('정수값을 입력하세요.: '))

recur(x)
````

````python
# 부분적으로 꼬리 재귀를 제거한 재귀 함수 구현하기(꼬리 재귀를 제거)

def recur(n: int) -> int:
    """꼬리 재귀를 제거한 recur() 함수"""
    while n > 0:
        recur(n - 1)
        print(n)
        n = n - 2

x = int(input('정수값을 입력하세요.: '))

recur(x)
````

````python
# while loop를 for loop로 표현

def recur(x: int) -> int:
    """순수한 재귀 함수 recur의 구현"""
    for n in range(x, 0, -2):
        recur(n - 1)
        print(n)

x = int(input('정수값을 입력하세요.: '))

recur(x)
````

위의 함수(`recur(n)`)를 수식(`R(n)`)으로 풀어쓰면 (점화식)

<br>

$$
\begin{aligned}
R(n)
& = R(n-1) + [n] \\
& + R(n-3) + [n-2] \\
& + R(n-5) + [n-4] \\
& + R(n-7) + [n-6] \\
& + \cdots \\
\\
& = R(n-1) + [n] \\
& + R(n-3) + [n-2] \\
& + R(n-5) + [n-4] \\
& + R(n-6)\\
& (\because\; R(n-6)=R(n-7)+[n-6]+\cdots)\\
\\
& = R(n-1) + [n] \\
& + R(n-3) + [n-2] \\
& + R(n-4) \\
& (\because\; R(n-4)=R(n-5)+[n-4]+R(n-6))\\
\\
& = R(n-1) + [n] \\
& + R(n-2) \\
& (\because\; R(n-2)=R(n-3)+[n-2]+R(n-4))\\
\\
\end{aligned}
$$

$$
R(n)=
\begin{cases}
[\;] & n \le 0 \\\\
R(n-1) + [n] + R(n-2) & n > 0
\end{cases}
$$

````python
def recur(n):
    if n <= 0:
        return []
    return recur(n-1) + [n] + recur(n-2)

n = 5

lst = recur(n)
for num in lst:
    print(num)
````
