````python
# 세 정수를 입력받아 최대값 구하기

print('세 정수의 최대값을 구합니다.')
a = int(input('정수 a의 값을 입력하세요.: '))
b = int(input('정수 b의 값을 입력하세요.: '))
c = int(input('정수 c의 값을 입력하세요.: '))

num_list = ['a', 'b', 'c']
max = a
num = 0
if max < b:
    max = b
    num = 1
if max < c:
    max = c
    num = 2

print(f'최댓값은 정수 {num_list[num]} {max}입니다.')
````

````python
# 세 정수를 입력받아 중앙값 구하기
def med3(a, b, c):
    if a > b:
        if c > a:
            return a
        elif b > c:
            return b
        else: # a > c, c > b
            return c
    elif c > b: # b > a
        return b
    elif a > c: # b > a, b > c
        return a
    else: # b > c, c > a
        return c
````

````python
# 가로, 세로 길이가 정수이고 넓이가 area인 직사각형에서 변의 길이 나열하기
import time
st = time.time()
area = 10**12

for i in range(1, area + 1): # 1부터 사각형 한 변 계산
    if i * i > area: # sqrt(n) 보다 작은 약수까지만 계산해주는 게 효율적
        break
    if area % i: # 약수 아니면 건너뛰기
        continue
    print(f'{i} x {area // i} = {i * (area // i)}')
time.time() - st
````

````python
import math
st = time.time()
area = 10**12
limit = math.isqrt(area)
for i in range(1, limit + 1): # 1부터 사각형 한 변 계산
    if area % i: # 약수 아니면 건너뛰기
        continue
    print(f'{i} x {area // i} = {i * (area // i)}')
time.time() - st
````

자연수 n이 합성수라면 
$$
n = a \cdot b \text{ \ }( 1 < a \leq b)
$$
일 때 반드시
$$
a \leq \sqrt{n}
$$
를 만족하는 **약수 a**가 존재합니다.

---
가정
\
모든 약수가 $\sqrt{n}$ 보다 크다고 가정:
$$
a \geq \sqrt{n}, \text{ } b \geq \sqrt{n}
$$
그러면
$$
ab \geq \sqrt{n} \cdot \sqrt{n} = n
$$
하지만 ab = n 이므로 **모순**.
\
$\therefore$ 합성수 n은 1을 제외한 $\sqrt{n}$ 이하의 약수를 적어도 하나 가진다.

---
$$\boxed{1 < a \le \sqrt{n} \le b}$$

합성수의 정의에 의해 
\
n = ab $( 1 < a \leq b)$인 약수 쌍이 존재한다.
\
이 때 $a \leq \sqrt{n}$이면 $b=\frac{n}{a}$ 가 항상 자동으로 결정되므로 
\
$\sqrt{n}$을 경계로 그 이하의 a값들만 탐색하는 게 효율적이다!

````python
# 10 ~ 99 사이의 난수 n개 생성하기(13이 나오면 중단)
import random

s = 13 # 중단할 수
n = 100 # 난수의 갯수

for _ in range(n):
    r = random.randint(1, 99)
    print(r)
    if r == s:
        print(f'\n {_+1}회 반복 후 {s}이 나왔으므로 프로그램을 중단합니다.')
        break
else:
    print(f'\n {_+1}회 반복 후 {s}이 안 나왔으므로 난수 생성을 종료합니다.')
````
