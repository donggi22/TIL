```python
# 세 정수를 입력받아 최대값 구하기

print('세 정수의 최대값을 구합니다.')
a = int(input('정수 a의 값을 입력하세요.: '))
b = int(input('정수 b의 값을 입력하세요.: '))
c = int(input('정수 c의 값을 입력하세요.: '))

num_list = ['a', 'b', 'c']
max = a
num = 0
if max < b:
    max = b
    num = 1
if max < c:
    max = c
    num = 2

print(f'최댓값은 정수 {num_list[num]} {max}입니다.')
```

```python
# 세 정수를 입력받아 중앙값 구하기
def med3(a, b, c):
    if a > b:
        if c > a:
            return a
        elif b > c:
            return b
        else: # a > c, c > b
            return c
    elif c > b: # b > a
        return b
    elif a > c: # b > a, b > c
        return a
    else: # b > c, c > a
        return c
```

```python
med3(1, 2, 3), med3(1, 3, 2), med3(2, 1, 3), med3(2, 3, 1), med3(3, 1, 2), med3(3, 2, 1)
```

```python
# 매우 비효율적으로 구하기
add = 0
i = 1
n = 10
num_list = list(range(1, n+1))
num_list[-1] = False
counter = True
while counter:
    add += i
    counter = num_list[i-1]
    i += 1
add
```

```python
# a부터 b까지 정수의 합 구하기 - 과정부터 결과까지 한 줄에 출력 ... + ... + ... = 결과
# a = int(input())
# b = int(input())

a, b = 15, 1

add = 0

if a > b:
    a, b = b, a

operator = ['+ ' for i in range(b)]
operator[-1] = '= '
for number in range(a, b + 1):
    add += number
    print(number, operator[number-1], end='')
print(add)
```

```python
a, b = 15, 1

if a > b:
    a, b = b, a

add = 0
for i in range(a, b + 1):
    add += i
    if i < b:
        print(f'{i} +', end=' ')
    else:
        print(f'{i} =', end=' ')
print(add)
```

```python
# +와 -를 번갈아 출력하기 -> 홀수: '-' 짝수: '+'

a, b = 1, 15
if a > b:
    a, b = b, a

for i in range(a, b+1):
    if i % 2 == 0:
        print('+', end='')
    else:
        print('-', end='')
```

```python
a, b = 2, 5
if a > b:
    a, b = b, a

if a & 1: # 홀수인지
    print('-', end='')

for i in range(a, (a + b + 1)//2):
    print('+-', end='')

if not(b & 1): # 짝수인지
    print('+')
```

```python
# *를 n개 출력하되 w개 마다 줄바꿈하기
n = 11
w = 5
for i in range(n):
    print('*', end='')
    if i % w == w - 1:
        print()

if n % w:
    print()
print('end')
```

```python
n = 11
w = 5
for i in range(n//w):
    print('*'*w)
rest = n % w
if rest:
    print('*'*rest)
print('end')
```
